## ğŸ¯ í•™ìŠµ ëª©í‘œ

- ë‘ ê°œ ì´ìƒì˜ ë³€ìˆ˜ ê°„ **ìƒê´€ê´€ê³„(Correlation)** ë¥¼ ì´í•´í•œë‹¤.  
- **ì‚°ì ë„(Scatter Plot)** ì™€ **ì‚°ì ë„ í–‰ë ¬(Scatter Matrix)** ë¡œ ê´€ê³„ë¥¼ ì‹œê°í™”í•œë‹¤.  
- ë²”ì£¼í˜• ë³€ìˆ˜ì— ë”°ë¼ ìƒ‰ì„ ë‹¬ë¦¬ í‘œí˜„í•´ **ê·¸ë£¹ë³„ ì°¨ì´**ë¥¼ í™•ì¸í•œë‹¤.  
- ì—¬ëŸ¬ ë³€ìˆ˜ ê°„ ê´€ê³„ë¥¼ ë™ì‹œì— ì‚´í´ë³´ëŠ” **seabornì˜ pairplot** ì„ í™œìš©í•œë‹¤.  
- ì‹œê³„ì—´/ë³µìˆ˜ ë³€ìˆ˜ì˜ **ì¶”ì„¸(Line Plot)** ë¥¼ ì´í•´í•œë‹¤.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1ï¸âƒ£ seabornì˜ ë‚´ì¥ ë°ì´í„° ì…‹ ë¶ˆëŸ¬ì˜¤ê¸° (ìë™ì°¨ ì—°ë¹„ ë°ì´í„°)
df = sns.load_dataset('mpg')  # 'mpg' â†’ miles per gallon (ì—°ë¹„)
df = df.dropna(subset=['weight', 'mpg'])  # ê²°ì¸¡ì¹˜(NaN) ì œê±°: weightë‚˜ mpgì— ê°’ì´ ì—†ëŠ” í–‰ ì‚­ì œ
```

```python
# 2ï¸âƒ£ ê¸°ë³¸ ì‚°ì ë„ (matplotlib ì‚¬ìš©)
plt.figure(figsize=(6, 4))                # ê·¸ë˜í”„ í¬ê¸° ì„¤ì • (ê°€ë¡œ 6, ì„¸ë¡œ 4)
plt.scatter(                             # ì‚°ì ë„(scatter plot) í•¨ìˆ˜
    df['weight'],                        # xì¶• â†’ ìë™ì°¨ ì¤‘ëŸ‰(weight)
    df['mpg'],                           # yì¶• â†’ ì—°ë¹„(mpg)
    color='teal'                         # ì  ìƒ‰ìƒ ì§€ì •
)
plt.title('Relationship between Weight and MPG')  # ê·¸ë˜í”„ ì œëª©
plt.xlabel('Weight (car weight)')                 # xì¶• ì´ë¦„
plt.ylabel('MPG (fuel efficiency)')               # yì¶• ì´ë¦„
plt.grid(True)                                   # ë°°ê²½ì— ê²©ìì„  í‘œì‹œ (ë°ì´í„° í•´ì„ ìš©ì´)
plt.show()                                       # ê·¸ë˜í”„ ì¶œë ¥
```

```python
# 3ï¸âƒ£ seabornì˜ scatterplot() í™œìš© (ìƒ‰ìƒ êµ¬ë¶„ + ìŠ¤íƒ€ì¼)
sns.scatterplot(
    data=df,                                    # ì‚¬ìš©í•  ë°ì´í„°í”„ë ˆì„
    x='weight',                                 # xì¶• ë³€ìˆ˜
    y='mpg',                                    # yì¶• ë³€ìˆ˜
    hue='origin',                               # ìƒ‰ìƒ ê¸°ì¤€ â†’ origin(ì œì¡°êµ­)
    style='origin',                             # ì  ëª¨ì–‘ ê¸°ì¤€ â†’ originë³„ë¡œ êµ¬ë¶„
    palette='Set2'                              # ìƒ‰ìƒ íŒ”ë ˆíŠ¸ ì§€ì • (ë³´ê¸° ì¢‹ì€ ê¸°ë³¸ í…Œë§ˆ)
)
plt.title('Relationship between Weight and MPG (by Origin)')  # ê·¸ë˜í”„ ì œëª©
plt.xlabel('Weight (car weight)')
plt.ylabel('MPG (fuel efficiency)')
plt.grid(True)
plt.show()
```

## ğŸ“Š 1. ì‚°ì ë„ (Scatter Plot)

- **ë‘ ì—°ì†í˜• ë³€ìˆ˜** ê°„ì˜ ê´€ê³„ë¥¼ ì‹œê°ì ìœ¼ë¡œ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.  
- ì ì˜ ìƒ‰(`hue`)ì´ë‚˜ ëª¨ì–‘(`style`)ìœ¼ë¡œ **ë²”ì£¼í˜• ë³€ìˆ˜**ë¥¼ êµ¬ë¶„í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.  
- ì˜ˆ: ìë™ì°¨ ì¤‘ëŸ‰ì´ í´ìˆ˜ë¡ ì—°ë¹„(`mpg`)ëŠ” ê°ì†Œí•˜ëŠ” ê²½í–¥ì„ ë³´ì„.

```python
# 1ï¸âƒ£ ì—¬ëŸ¬ ë³€ìˆ˜ ê°„ì˜ ê´€ê³„ë¥¼ ë™ì‹œì— ì‹œê°í™” (Pairplot)
# â†’ ê° ë³€ìˆ˜ ìŒì— ëŒ€í•´ ì‚°ì ë„(scatter plot)ë¥¼ ìë™ìœ¼ë¡œ ìƒì„±í•˜ê³ ,
#   ëŒ€ê°ì„ (diagonal)ì—ëŠ” íˆìŠ¤í† ê·¸ë¨(histogram) ë˜ëŠ” ì»¤ë„ ë°€ë„ ê·¸ë˜í”„(kde)ë¥¼ í‘œì‹œí•¨
sns.pairplot(
    df[['mpg', 'horsepower', 'weight', 'acceleration', 'origin']],  # ë¶„ì„í•  ë³€ìˆ˜ ì„ íƒ
    hue='origin',        # ë²”ì£¼í˜• ë³€ìˆ˜(ì œì¡°êµ­)ì— ë”°ë¼ ìƒ‰ìƒ êµ¬ë¶„
    palette='coolwarm'   # ìƒ‰ìƒ íŒ”ë ˆíŠ¸ ì§€ì • (ë¶‰ì€ìƒ‰~í‘¸ë¥¸ìƒ‰ ê³„ì—´)
)

# 2ï¸âƒ£ ì „ì²´ ê·¸ë˜í”„ ì œëª© ì„¤ì •
plt.suptitle('Relationships among Multiple Variables (MPG, Horsepower, Weight, Acceleration)',
             y=1.02)  # y=1.02 â†’ ì œëª©ì´ ê·¸ë˜í”„ì— ê²¹ì¹˜ì§€ ì•Šë„ë¡ ì•½ê°„ ìœ„ë¡œ ì´ë™
plt.show()
```

## ğŸ§© 2. ë‹¤ì¤‘ ì‚°ì ë„ í–‰ë ¬ (Pairplot)

- ì—¬ëŸ¬ ë³€ìˆ˜ ê°„ **ìƒí˜¸ ê´€ê³„**ë¥¼ í•œëˆˆì— í™•ì¸í•  ìˆ˜ ìˆëŠ” ê·¸ë˜í”„ì…ë‹ˆë‹¤.  
- ëŒ€ê°ì„ ì—ëŠ” ê° ë³€ìˆ˜ì˜ **ë¶„í¬(íˆìŠ¤í† ê·¸ë¨)**,  
  ë‚˜ë¨¸ì§€ì—ëŠ” **ì‚°ì ë„(Scatter Plot)** ê°€ í‘œì‹œë©ë‹ˆë‹¤.  
- ì˜ˆ: `mpg`(ì—°ë¹„)ì™€ `weight`ëŠ” ìŒì˜ ìƒê´€ê´€ê³„,  
  `horsepower`ì™€ `weight`ëŠ” ì–‘ì˜ ìƒê´€ê´€ê³„ë¥¼ ë³´ì„.

```python
# 1ï¸âƒ£ Seabornì˜ ë‚´ì¥ ë°ì´í„° ì…‹ ë¶ˆëŸ¬ì˜¤ê¸° (Iris ë°ì´í„°)
iris = sns.load_dataset('iris')
# iris: ì„¸ ê°€ì§€ í’ˆì¢…(setosa, versicolor, virginica)ì˜ ê½ƒë°›ì¹¨(sepal)ê³¼ ê½ƒì(petal) ê¸¸ì´Â·ë„ˆë¹„ ë°ì´í„°

# 2ï¸âƒ£ í’ˆì¢…(species)ë³„ë¡œ ê½ƒìì˜ ê¸¸ì´(petal_length)ì™€ í­(petal_width) ê´€ê³„ë¥¼ ì‹œê°í™”
plt.figure(figsize=(6, 4))  # ê·¸ë˜í”„ í¬ê¸° ì„¤ì • (ê°€ë¡œ 6, ì„¸ë¡œ 4)

sns.scatterplot(
    data=iris,                # ì‚¬ìš©í•  ë°ì´í„°í”„ë ˆì„
    x='petal_length',         # xì¶• ë³€ìˆ˜ â†’ ê½ƒì ê¸¸ì´
    y='petal_width',          # yì¶• ë³€ìˆ˜ â†’ ê½ƒì í­
    hue='species',            # ìƒ‰ìƒ êµ¬ë¶„ ê¸°ì¤€ â†’ í’ˆì¢…ë³„ êµ¬ë¶„
    style='species',          # ì  ëª¨ì–‘(style)ë„ í’ˆì¢…ë³„ë¡œ ë‹¤ë¥´ê²Œ ì§€ì •
    s=80                      # ì ì˜ í¬ê¸° (ê¸°ë³¸ë³´ë‹¤ ì•½ê°„ í¬ê²Œ ì„¤ì •)
)

# 3ï¸âƒ£ ê·¸ë˜í”„ ì œëª© ë° ì¶• ì´ë¦„ ì„¤ì •
plt.title('Relationship between Petal Length and Width by Species')  # ì œëª©
plt.xlabel('Petal Length (cm)')  # xì¶• ì´ë¦„
plt.ylabel('Petal Width (cm)')   # yì¶• ì´ë¦„
plt.grid(True, linestyle='--', alpha=0.5)  # ë°°ê²½ ê²©ìì„  ì¶”ê°€ (ê°€ë…ì„± í–¥ìƒ)
plt.show()
```

## ğŸŒº 3. ê·¸ë£¹ë³„ ì‚°ì ë„ (Scatter Plot by Group)

- `iris` ë°ì´í„° ì…‹ì€ ëŒ€í‘œì ì¸ ë‹¤ì¤‘ ë³€ìˆ˜ ì˜ˆì‹œì…ë‹ˆë‹¤.  
- í’ˆì¢…(`species`)ì— ë”°ë¼ **ê½ƒìì˜ ê¸¸ì´(`petal_length`)** ì™€  
  **í­(`petal_width`)** ì˜ ê´€ê³„ê°€ ë‹¤ë¥´ê²Œ ë‚˜íƒ€ë‚©ë‹ˆë‹¤.  
- ìƒ‰ìƒ êµ¬ë¶„ì„ í†µí•´ ê·¸ë£¹ ê°„ ì°¨ì´ë¥¼ ì§ê´€ì ìœ¼ë¡œ íŒŒì•…í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
# 1ï¸âƒ£ Iris ë°ì´í„° ì…‹ì˜ ìƒê´€ê³„ìˆ˜(correlation coefficient) ê³„ì‚°
# â†’ ìƒê´€ê³„ìˆ˜: ë‘ ë³€ìˆ˜ ê°„ì˜ ê´€ê³„ ê°•ë„ì™€ ë°©í–¥ì„ ë‚˜íƒ€ë‚´ëŠ” ê°’ (-1 ~ +1)
iris_corr = iris.corr(numeric_only=True)  # ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ë§Œ ì„ íƒí•´ì„œ ìƒê´€ê´€ê³„ ê³„ì‚°
print(iris_corr)  # ê³„ì‚°ëœ ìƒê´€ê³„ìˆ˜ í–‰ë ¬ ì¶œë ¥
```

```python
# 2ï¸âƒ£ Heatmapìœ¼ë¡œ ì‹œê°í™”
plt.figure(figsize=(6, 5))  # ê·¸ë˜í”„ í¬ê¸° ì„¤ì •

sns.heatmap(
    iris_corr,               # ìƒê´€ê³„ìˆ˜ í–‰ë ¬
    annot=True,              # ì…€ ì•ˆì— ê°’ í‘œì‹œ
    cmap='coolwarm',         # ìƒ‰ìƒ íŒ”ë ˆíŠ¸ (ë¶‰ì€ìƒ‰=ì–‘ì˜ ìƒê´€, í‘¸ë¥¸ìƒ‰=ìŒì˜ ìƒê´€)
    fmt=".2f"                # ì†Œìˆ˜ì  ë‘˜ì§¸ ìë¦¬ê¹Œì§€ í‘œì‹œ
)

# 3ï¸âƒ£ ì œëª© ì„¤ì •
plt.title('Correlation Heatmap of Iris Variables')  # ê·¸ë˜í”„ ì œëª©
plt.show()
```

## ğŸ”— 4. ìƒê´€ê³„ìˆ˜ (Correlation)

- ë³€ìˆ˜ ê°„ì˜ **ì„ í˜• ê´€ê³„ ì •ë„**ë¥¼ ìˆ˜ì¹˜ë¡œ í‘œí˜„í•©ë‹ˆë‹¤.  
- `+1`: ê°•í•œ ì–‘ì˜ ìƒê´€ê´€ê³„ (í•¨ê»˜ ì¦ê°€)  
- `0`: ê´€ê³„ ì—†ìŒ  
- `-1`: ê°•í•œ ìŒì˜ ìƒê´€ê´€ê³„ (í•œìª½ì´ ì¦ê°€í•˜ë©´ ë‹¤ë¥¸ ìª½ ê°ì†Œ)  
- ì˜ˆ: ê½ƒì ê¸¸ì´(`petal_length`)ì™€ í­(`petal_width`)ì˜ ìƒê´€ê³„ìˆ˜ â‰ˆ 0.96 â†’ ë§¤ìš° ê°•í•œ ì–‘ì˜ ìƒê´€.

```python
# 1ï¸âƒ£ seabornì˜ ë‚´ì¥ ë°ì´í„° ì…‹ 'tips' ë¶ˆëŸ¬ì˜¤ê¸°
tips = sns.load_dataset('tips')  # ì‹ë‹¹ì—ì„œ íŒì„ ì¤€ ë°ì´í„° (ìš”ì¼, ì„±ë³„, ê¸ˆì•¡ ë“± í¬í•¨)

# 2ï¸âƒ£ ìš”ì¼ë³„ í‰ê·  íŒ ë¹„ìœ¨ ê³„ì‚°
#    â†’ íŒ ë¹„ìœ¨ = (í‰ê·  íŒ ê¸ˆì•¡ / í‰ê·  ì´ ì§€ë¶ˆ ê¸ˆì•¡)
tip_rate = tips.groupby('day', observed=True)['tip'].mean() / \
           tips.groupby('day', observed=True)['total_bill'].mean()

# 3ï¸âƒ£ ì„ ê·¸ë˜í”„ë¡œ ìš”ì¼ë³„ íŒ ë¹„ìœ¨ ì‹œê°í™”
plt.plot(
    tip_rate.index,         # xì¶•: ìš”ì¼ (Thu, Fri, Sat, Sun)
    tip_rate.values,        # yì¶•: íŒ ë¹„ìœ¨
    marker='o',             # ë°ì´í„° í¬ì¸íŠ¸ í‘œì‹œ (ì›í˜• ë§ˆì»¤)
    linestyle='-',          # ì„  ìŠ¤íƒ€ì¼ (ì‹¤ì„ )
    color='purple'          # ì„  ìƒ‰ìƒ
)

# 4ï¸âƒ£ ê·¸ë˜í”„ ì œëª©ê³¼ ì¶• ë ˆì´ë¸” ì„¤ì •
plt.title('Average Tip Rate by Day')     # ê·¸ë˜í”„ ì œëª©
plt.xlabel('Day')                        # xì¶• ë ˆì´ë¸”
plt.ylabel('Tip Rate (tip / total_bill)') # yì¶• ë ˆì´ë¸”

# 5ï¸âƒ£ ê²©ì ì¶”ê°€ (ê°€ë…ì„± í–¥ìƒ)
plt.grid(True)

# 6ï¸âƒ£ ê·¸ë˜í”„ ì¶œë ¥
plt.show()
```

## ğŸ“ˆ 5. ì„ ê·¸ë˜í”„ (Line Plot)

- ì‹œê³„ì—´ ë°ì´í„°ë‚˜ ë³€í™” ì¶”ì„¸ë¥¼ í‘œí˜„í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.  
- `tips` ë°ì´í„°ì—ì„œ ìš”ì¼ë³„ í‰ê·  íŒ ë¹„ìœ¨(`tip_rate`)ì˜ ë³€í™” ì¶”ì„¸ë¥¼ í™•ì¸í•©ë‹ˆë‹¤.  
- ì˜ˆ: ì£¼ì¤‘(ëª©Â·ê¸ˆ)ì—ëŠ” í‰ê·  íŒ ë¹„ìœ¨ì´ ë‹¤ì†Œ ë†’ê²Œ ë‚˜íƒ€ë‚¨.

## ğŸ“Š ë°ì´í„° ê°œìš”

- **mpg** ë°ì´í„° ì…‹ì€ ìë™ì°¨ì˜ ì—°ë¹„, ë°°ê¸°ëŸ‰, ì¤‘ëŸ‰, ë§ˆë ¥ ë“±ì˜ ì •ë³´ë¥¼ ë‹´ê³  ìˆìŒ.  
- ì£¼ìš” ë³€ìˆ˜:
  - **mpg**: ì—°ë¹„ (ì—°ì†í˜•)
  - **horsepower**: ë§ˆë ¥
  - **weight**: ì°¨ëŸ‰ ì¤‘ëŸ‰
  - **acceleration**: ê°€ì†ë„
  - **origin**: ì œì¡° ì§€ì—­ (categorical)

ì´ë²ˆ ì‹¤ìŠµì—ì„œëŠ”  
â‘  ë³€ìˆ˜ ê°„ì˜ ê´€ê³„ë¥¼ ì‚°ì ë„ë¡œ ì‹œê°í™”í•˜ê³   
â‘¡ ìƒê´€ê´€ê³„ë¥¼ ê³„ì‚°í•˜ë©°  
â‘¢ ì œì¡° ì§€ì—­ë³„ í‰ê·  ë¹„êµë¥¼ ìˆ˜í–‰í•©ë‹ˆë‹¤.

```python
import seaborn as sns
import pandas as pd

# ë‚´ì¥ ë°ì´í„° ì…‹: ë¯¸êµ­ ì£¼ë³„ ì‚¬íšŒê²½ì œ ì •ë³´ ëŒ€ì²´ìš©
# seabornì—ëŠ” state_x77ì™€ ìœ ì‚¬í•œ 'mpg' ë°ì´í„° ì…‹ ì‚¬ìš©
df = sns.load_dataset('mpg')
df.head()
```

## ğŸ“ˆ ì‚°ì ë„(Scatter Plot) í•´ì„

- **ë°°ê¸°ëŸ‰ì´ í´ìˆ˜ë¡ ì—°ë¹„ëŠ” ë‚®ìŒ(ìŒì˜ ìƒê´€ê´€ê³„)**  
- ìƒê´€ê³„ìˆ˜ âˆ’1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ **ì„ í˜• ê´€ê³„ê°€ ê°•í•¨**  
  â†’ `-0.8 ì´ìƒ`ì´ë©´ ê°•í•œ ìŒì˜ ìƒê´€ê´€ê³„ë¡œ í•´ì„ ê°€ëŠ¥

```python
# ë°°ê¸°ëŸ‰(displacement)ê³¼ ì—°ë¹„(mpg)ì˜ ê´€ê³„ ì‹œê°í™” ë° ìƒê´€ë¶„ì„
# ğŸš— ë°ì´í„°ëŠ” ì´ì „ì— sns.load_dataset('mpg') ë¡œ ë¶ˆëŸ¬ì˜¨ df ì‚¬ìš©

# 1ï¸âƒ£ ì‚°ì ë„ (Scatter Plot)
df.plot.scatter(
    x='displacement',        # xì¶•: ì—”ì§„ ë°°ê¸°ëŸ‰
    y='mpg',                 # yì¶•: ì—°ë¹„
    color='steelblue',       # ì  ìƒ‰ìƒ
    alpha=0.7                # ì ì˜ íˆ¬ëª…ë„ (1.0ì€ ë¶ˆíˆ¬ëª…, 0ì€ ì™„ì „ íˆ¬ëª…)
)
plt.title('Relationship between Engine Displacement and MPG')
plt.xlabel('Displacement (Engine Size)')
plt.ylabel('MPG (Fuel Efficiency)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# 2ï¸âƒ£ ìƒê´€ê³„ìˆ˜ ê³„ì‚° (Correlation Coefficient)
# corr() í•¨ìˆ˜ëŠ” ë‘ ë³€ìˆ˜ ê°„ì˜ ìƒê´€ê´€ê³„ë¥¼ ê³„ì‚°
# ê²°ê³¼ê°’: -1 ~ +1 ì‚¬ì´ (ìŒìˆ˜ = ë°˜ë¹„ë¡€, ì–‘ìˆ˜ = ë¹„ë¡€, 0 = ê´€ê³„ ì—†ìŒ)
corr_value = df['displacement'].corr(df['mpg'])

# ì†Œìˆ˜ì  3ìë¦¬ë¡œ í‘œì‹œ
print(f"Correlation between displacement and mpg: {corr_value:.3f}")
```

## ğŸ” ì‚°ì ë„ í–‰ë ¬(Scatter Matrix)

- ì—¬ëŸ¬ ë³€ìˆ˜ ê°„ ê´€ê³„ë¥¼ í•œëˆˆì— íŒŒì•…í•  ìˆ˜ ìˆìŒ  
- ëŒ€ê°ì„ (diagonal)ì€ íˆìŠ¤í† ê·¸ë¨ â†’ ê° ë³€ìˆ˜ì˜ ë¶„í¬  
- ë‚˜ë¨¸ì§€ ì¹¸ì€ ë³€ìˆ˜ ê°„ ì‚°ì ë„ â†’ ê´€ê³„ í™•ì¸  
  - ì˜ˆ: **ë¬´ê²Œ(weight)ê°€ ì¦ê°€í•˜ë©´ ì—°ë¹„(mpg)ëŠ” ê°ì†Œ**

```python
import pandas as pd
import matplotlib.pyplot as plt

# âœ… ë‹¤ì¤‘ ë³€ìˆ˜ì˜ ê´€ê³„ë¥¼ í•œëˆˆì— ë³´ê¸° ìœ„í•œ "ì‚°ì ë„ í–‰ë ¬(Scatter Matrix)"
# ì—¬ëŸ¬ ë³€ìˆ˜ ê°„ ìƒê´€ê´€ê³„ì™€ ë¶„í¬ë¥¼ ë™ì‹œì— ì‹œê°í™”í•  ìˆ˜ ìˆìŒ

pd.plotting.scatter_matrix(
    df[['mpg', 'horsepower', 'weight', 'acceleration']],  # ë¶„ì„í•  ë³€ìˆ˜ 4ê°œ ì„ íƒ
    figsize=(10, 10),       # ê·¸ë˜í”„ í¬ê¸° ì„¤ì • (inch ë‹¨ìœ„)
    diagonal='hist',        # ëŒ€ê°ì„  ë°©í–¥ì—ëŠ” íˆìŠ¤í† ê·¸ë¨(histogram) í‘œì‹œ
    color='teal'            # ì‚°ì ë„ ì  ìƒ‰ìƒ ì§€ì •
)

# ì „ì²´ ì œëª© ì„¤ì •
plt.suptitle('Scatter Matrix: Relationships among Multiple Variables', fontsize=14)
plt.show()
```

## ğŸ“ ìƒê´€ê³„ìˆ˜ í–‰ë ¬(Correlation Matrix)

| ê°’ ë²”ìœ„ | í•´ì„ |
|----------|------|
| **+1** | ì™„ë²½í•œ ì–‘ì˜ ìƒê´€ (ê°™ì´ ì¦ê°€) |
| **0** | ìƒê´€ ì—†ìŒ |
| **âˆ’1** | ì™„ë²½í•œ ìŒì˜ ìƒê´€ (í•˜ë‚˜ ì¦ê°€ â†’ ë‹¤ë¥¸ ê°ì†Œ) |

- `weight` â†” `mpg` : ê°•í•œ ìŒì˜ ìƒê´€  
- `horsepower` â†” `mpg` : ìŒì˜ ìƒê´€  
- `acceleration` â†” `mpg` : ì•½í•œ ì–‘ì˜ ìƒê´€

```python
# âœ… ìƒê´€ê³„ìˆ˜ í–‰ë ¬(correlation matrix) ê³„ì‚°
# ê° ë³€ìˆ˜ ê°„ì˜ ì„ í˜• ê´€ê³„(ìƒê´€ì„±)ë¥¼ ìˆ˜ì¹˜ë¡œ í‘œí˜„ (ê°’ì˜ ë²”ìœ„: -1 ~ +1)

corr_matrix = df[['mpg', 'horsepower', 'weight', 'acceleration']].corr()  # ì„ íƒí•œ ë³€ìˆ˜ ê°„ ìƒê´€ê³„ìˆ˜ ê³„ì‚°

print("ë³€ìˆ˜ ê°„ ìƒê´€ê³„ìˆ˜ í–‰ë ¬:")
print(corr_matrix)
```

## ğŸŒ ê·¸ë£¹ë³„ ë¹„êµ ë¶„ì„

- `groupby()`ë¥¼ í†µí•´ ì§€ì—­ë³„(Origin) í‰ê·  ì—°ë¹„ ë¹„êµ  
- **ê²°ê³¼:**  
  - `Japan` ìë™ì°¨ â†’ í‰ê·  ì—°ë¹„ ê°€ì¥ ë†’ìŒ  
  - `USA` ìë™ì°¨ â†’ ê°€ì¥ ë‚®ìŒ  
- ì§€ì—­ë³„ë¡œ ì œì¡° íŠ¹ì„± ì°¨ì´ë¥¼ ë°˜ì˜í•œ ê²°ê³¼  
  (ì—”ì§„ íš¨ìœ¨, ì°¨ëŸ‰ í¬ê¸° ë“±)

```python
# âœ… 1ï¸âƒ£ ì œì¡° ì§€ì—­(origin)ë³„ í‰ê·  ì—°ë¹„ ê³„ì‚°
# groupby('origin'): origin(ì œì¡° ì§€ì—­)ë³„ë¡œ ë°ì´í„°ë¥¼ ê·¸ë£¹í™”
# ['mpg'].mean(): ê° ê·¸ë£¹ ë‚´ mpg(ì—°ë¹„)ì˜ í‰ê· ì„ ê³„ì‚°
# sort_values(ascending=False): í‰ê·  ì—°ë¹„ë¥¼ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬
region_mean = df.groupby('origin')['mpg'].mean().sort_values(ascending=False)
print("ì§€ì—­ë³„ í‰ê·  ì—°ë¹„:\n", region_mean, "\n")
```

```python
# âœ… 2ï¸âƒ£ ì‹œê°í™” - ë§‰ëŒ€ê·¸ë˜í”„(bar chart)
region_mean.plot.bar(
    color=['orange', 'lightblue', 'green']  # ì§€ì—­ë³„ë¡œ êµ¬ë¶„ëœ ìƒ‰ìƒ ì§€ì •
)
plt.title('Average MPG by Region of Manufacture')  # ê·¸ë˜í”„ ì œëª©
plt.xlabel('Region of Manufacture (origin)')        # xì¶• ë ˆì´ë¸”
plt.ylabel('Average MPG')                           # yì¶• ë ˆì´ë¸”
plt.grid(True, axis='y', linestyle='--', alpha=0.5) # yì¶• ë°©í–¥ ì ì„  ê²©ì ì¶”ê°€
plt.show()
```
